```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
from skimage import io as io_url
```

```python
frame = io_url.imread('https://raw.githubusercontent.com/moaathalrajab/CSC467_Week2_Lab/refs/heads/main/Lena.png')
frame = np.float32(frame) / 255

plt.imshow(frame)
```

```python
print(frame.shape)
```

```python
frm= frame[0:100,0:100,0]
plt.imshow(frm, cmap="gray")
```

```python
new_array= frm+20
plt.imshow(new_array, cmap='gray')
```

```python
kernel = (5, 5)
mean_frame_lena = cv2.blur(frame, kernel)
fig, ax = plt.subplots(1,1)
ax.imshow(mean_frame_lena)
```

```python
sharpen_matrix = np.array([[0, -1, 0], [-1,  5, -1], [0, -1, 0]], dtype=np.float32)

sharpened_lena = cv2.filter2D(frame, -1, sharpen_matrix)

fig, ax = plt.subplots(1,1)
ax.imshow(sharpened_lena)
```

```python
edges_lena = cv2.Canny((frame * 255).astype(np.uint8), threshold1=100, threshold2=200)
fix, ax = plt.subplots(1,1)
ax.imshow(edges_lena)
```
